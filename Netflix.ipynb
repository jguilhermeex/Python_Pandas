
## Preparando os dados

- Retirando colunas irrelevantes
- Identificar a existência de linhas duplicadas
- Renomear colunas para melhor entendimento
- Tratamento de valores ausentes (missing values).

```python
#Removendo a coluna Unnamed
netflix.drop(['Unnamed: 0'], axis = 1, inplace = True)
netflix

```

```python
# Checando linhas duplicadas
netflix.duplicated()
```

```python
# Checando linhas duplicadas
netflix.duplicated().sum()
```

```python
#Renomeando a coluna listed_in
netflix.rename(columns = {'listed_in' : 'categories'}, inplace=True)
netflix

```

```python
#Verificando os valores ausentes
netflix.isnull().sum()
```

```python
#Removi os valores ausentes das colunas rating, duration e date_added pois possuiam quantidades irrelevantes.
netflix.dropna(subset = ['rating','duration','date_added'], inplace = True)
```

```python
netflix.isnull().sum()
```

```python
#Verificando a porcentagem de valores nulos
netflix.isnull().sum() / netflix.shape[0] * 100
#A coluna director possui um alto índice de valores ausentes,no entanto possui uma informação relevante.
#Preenchi as colunas cast e country com Unknown para não enviesar os dados.
```

```python
#Preenchendo os valores nulos com Unknown nas colunas cast e country e removendo a coluna director
netflix['cast'].fillna(value = 'Unknown', inplace = True)
netflix['country'].fillna(value = 'Unknown', inplace = True)
netflix.drop(columns =['director'], inplace = True)
netflix.isnull().sum()
```

```python
#checando novamente os dados
netflix.info()
```

# Visualizando os dados

```python
#Analisei o numero de produções por país e os Estados Unidos aparece bem a frente de todos os outros
ax = netflix['country'].value_counts().head(10).plot(figsize =(15,5),\
title ='Número de Países Com Mais Produções da Netflix',kind ='bar', color = 'green')
ax.set_xlabel('Países')
ax.set_ylabel('Contagem')
```

```python
#Existem mais filmes do que séries no catálogo da netflix
colors = ['#0089ba','#f9f871']
netflix['type'].value_counts().plot(figsize =(20,6),\
title ='Porcentagem de Séries e Filmes no catálogo da Netflix',kind = 'pie', \
legend =True, autopct='%1.1f%%', ylabel ='', colors=colors)


```
